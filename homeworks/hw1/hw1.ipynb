Let $\theta = (\theta_0, \dots, \theta_{d-1}) \in \mathbb{R}^{d}$ and define the model $p_\theta(x) = \frac{e^{\theta_x}}{\sum_{x'}e^{\theta_{x'}}}$

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import torch.nn.functional as F
import math
from torch.utils.data import DataLoader, TensorDataset
```

```python
def q1_a(train_data, test_data, d, dset_id):
  """
  train_data: An (n_train,) numpy array of integers in {0, ..., d-1}
  test_data: An (n_test,) numpy array of integers in {0, .., d-1}
  d: The number of possible discrete values for random variable x
  dset_id: An identifying number of which dataset is given (1 or 2). Most likely
             used to set different hyperparameters for different datasets

  Returns
  - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
  - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
  - a numpy array of size (d,) of model probabilities
  """

  learning_rate = 0.1 if dset_id == 1 else 0.05
  batch_size = 32
  epochs = 50

  # Convert data to PyTorch tensors
  train_data = torch.tensor(train_data, dtype=torch.long)
  test_data = torch.tensor(test_data, dtype=torch.long)

  # Model: Categorical Distribution Parameters
  class CategoricalModel(nn.Module):
      def __init__(self, d):
          super().__init__()
          self.theta = nn.Parameter(torch.zeros(d))

      def forward(self, x):
          # Softmax to ensure probabilities sum to 1
          probs = torch.softmax(self.theta, dim=0)

          # Negative log-likelihood for categorical distribution
          log_probs = torch.log_softmax(self.theta, dim=0)
          nll = -log_probs[x]
          return nll.mean()

  # Initialize model and optimizer
  model = CategoricalModel(d)
  optimizer = optim.SGD(model.parameters(), lr=learning_rate)

  # Tracking losses
  train_losses = []
  test_losses = []

  # Initial test loss
  with torch.no_grad():
      test_loss = torch.mean(torch.stack([model(x) for x in test_data]))
      test_losses.append(test_loss.item())

  # Training loop
  for epoch in range(epochs):
      # Shuffle training data
      perm = torch.randperm(len(train_data))
      train_data_shuffled = train_data[perm]

      epoch_train_losses = []
      for i in range(0, len(train_data), batch_size):
          batch = train_data_shuffled[i:i+batch_size]

          # Zero gradients
          optimizer.zero_grad()

          # Compute loss
          loss = model(batch)
          loss.backward()

          # Update parameters
          optimizer.step()

          epoch_train_losses.append(loss.item())

      # Average train loss for the epoch
      train_losses.extend(epoch_train_losses)

      # Compute test loss after epoch
      with torch.no_grad():
          test_loss = torch.mean(torch.stack([model(x) for x in test_data]))
          test_losses.append(test_loss.item())

  # Final model probabilities
  with torch.no_grad():
      distribution = torch.softmax(model.theta, dim=0).numpy()

  # Convert to numpy arrays
  train_losses = np.array(train_losses)
  test_losses = np.array(test_losses)

  return train_losses, test_losses, distribution
```

## Part (b) Fitting Discretized Mixture of Logistics

Let us model $p_\theta(x)$ as a **discretized** mixture of 4 logistics such that $p_\theta(x) = \sum_{i=1}^4 \pi_i[\sigma((x+0.5 - \mu_i)/s_i) - \sigma((x-0.5-\mu_i)/s_i)]$

For the edge case of when $x = 0$, we replace $x-0.5$ by $-\infty$, and for $x = 99$, we replace $x+0.5$ by $\infty$.

You may find the [PixelCNN++](https://arxiv.org/abs/1701.05517) helpful for more information on discretized mixture of logistics.

**Provide the same set of corresponding deliverables as part (a)**

Fill out the function below and return the necessary arguments. Feel free to create more cells if need be.

```python
def q1_b(train_data, test_data, d, dset_id):
  """
  train_data: An (n_train,) numpy array of integers in {0, ..., d-1}
  test_data: An (n_test,) numpy array of integers in {0, .., d-1}
  d: The number of possible discrete values for random variable x
  dset_id: An identifying number of which dataset is given (1 or 2). Most likely
           used to set different hyperparameters for different datasets

  Returns
  - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
  - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
  - a numpy array of size (d,) of model probabilities
  """
  learning_rate = 0.001 if dset_id == 1 else 0.0005
  batch_size = 64
  epochs = 100
  num_mixtures = 4

  # Convert data to PyTorch tensors
  train_data = torch.tensor(train_data, dtype=torch.float32)
  test_data = torch.tensor(test_data, dtype=torch.float32)

  class DiscretizedLogisticsMixture(nn.Module):
      def __init__(self, d, num_mixtures):
          super().__init__()
          # Mixture weights (log-softmax to ensure positivity)
          self.mixture_weights = nn.Parameter(torch.zeros(num_mixtures))

          # Logistics parameters
          self.means = nn.Parameter(torch.linspace(0, d-1, num_mixtures))
          self.log_scales = nn.Parameter(torch.zeros(num_mixtures))

      def discretized_logistic_cdf(self, x, mu, s):
          # Implements CDF of logistic distribution with safe handling for edges
          x = torch.clamp(x, 0, d-1)
          return torch.sigmoid((x - mu) / s)

      def log_prob(self, x):
          # Mixture of discretized logistics log probability
          weights = F.log_softmax(self.mixture_weights, dim=0)

          log_probs = []
          for i in range(len(self.means)):
              mu = self.means[i]
              s = torch.exp(self.log_scales[i])

              # Handle edge cases
              cdf_upper = torch.where(x == d-1,
                                      torch.tensor(1.0).to(x.device),
                                      self.discretized_logistic_cdf(x + 0.5, mu, s))
              cdf_lower = torch.where(x == 0,
                                      torch.tensor(0.0).to(x.device),
                                      self.discretized_logistic_cdf(x - 0.5, mu, s))

              # Compute log probability for this mixture component
              log_prob_component = torch.log(torch.clamp(cdf_upper - cdf_lower, min=1e-10))
              log_probs.append(weights[i] + log_prob_component)

          # Logsumexp to combine mixture components
          return torch.logsumexp(torch.stack(log_probs), dim=0)

      def forward(self, x):
          # Negative log-likelihood
          return -torch.mean(self.log_prob(x))

  # Initialize model and optimizer
  model = DiscretizedLogisticsMixture(d, num_mixtures)
  optimizer = optim.Adam(model.parameters(), lr=learning_rate)

  # Tracking losses
  train_losses = []
  test_losses = []

  # Initial test loss
  with torch.no_grad():
      test_loss = model(test_data)
      test_losses.append(test_loss.item())

  # Training loop
  for epoch in range(epochs):
      # Shuffle training data
      perm = torch.randperm(len(train_data))
      train_data_shuffled = train_data[perm]

      epoch_train_losses = []
      for i in range(0, len(train_data), batch_size):
          batch = train_data_shuffled[i:i+batch_size]

          # Zero gradients
          optimizer.zero_grad()

          # Compute loss
          loss = model(batch)
          loss.backward()

          # Update parameters
          optimizer.step()

          epoch_train_losses.append(loss.item())

      # Average train loss for the epoch
      train_losses.extend(epoch_train_losses)

      # Compute test loss after epoch
      with torch.no_grad():
          test_loss = model(test_data)
          test_losses.append(test_loss.item())

  # Compute final distribution probabilities
  with torch.no_grad():
      # Compute probability for each discrete value
      x_range = torch.arange(d, dtype=torch.float32)
      distribution = torch.exp(model.log_prob(x_range)).numpy()
      distribution /= distribution.sum()  # Normalize

  # Convert to numpy arrays
  train_losses = np.array(train_losses)
  test_losses = np.array(test_losses)

  return train_losses, test_losses, distribution
```

```python
class MaskedConv2d(nn.Conv2d):
    def __init__(self, mask_type, *args, **kwargs):
        super().__init__(*args, **kwargs)
        assert mask_type in ['A', 'B'], "Mask type must be 'A' or 'B'"
        self.register_buffer('mask', torch.ones_like(self.weight))

        # Create mask
        C_out, C_in, H, W = self.weight.shape
        center_h, center_w = H // 2, W // 2

        # Mask out future pixels
        self.mask[:, :, center_h+1:, :] = 0
        self.mask[:, :, center_h, center_w+1:] = 0

        # Special handling for type A mask
        if mask_type == 'A':
            self.mask[:, :, center_h, center_w] = 0

    def forward(self, x):
        # Apply mask to weights before convolution
        masked_weight = self.weight * self.mask
        return F.conv2d(x, masked_weight, self.bias,
                        self.stride, self.padding,
                        self.dilation, self.groups)

class PixelCNN(nn.Module):
    def __init__(self, image_shape, n_filters=64):
        super().__init__()
        H, W = image_shape

        # Initial masked type A convolution
        self.first_conv = MaskedConv2d('A', 1, n_filters, 7, padding=3)

        # 5 masked type B convolutions
        self.type_b_convs1 = nn.ModuleList([
            nn.Sequential(
                MaskedConv2d('B', n_filters, n_filters, 7, padding=3),
                nn.ReLU()
            ) for _ in range(5)
        ])

        # 2 1x1 masked type B convolutions
        self.type_b_convs2 = nn.ModuleList([
            nn.Sequential(
                MaskedConv2d('B', n_filters, n_filters, 1),
                nn.ReLU()
            ) for _ in range(2)
        ])

        # Final convolution to output logits
        self.final_conv = nn.Conv2d(n_filters, 1, 1)

    def forward(self, x):
        # Initial convolution
        x = self.first_conv(x)

        # Type B convolutions
        for conv in self.type_b_convs1:
            x = conv(x)

        # 1x1 convolutions
        for conv in self.type_b_convs2:
            x = conv(x)

        # Final convolution to logits
        return self.final_conv(x)
```

```python
def q2_a(train_data, test_data, image_shape, dset_id):
  """
  train_data: A (n_train, H, W, 1) uint8 numpy array of binary images with values in {0, 1}
  test_data: A (n_test, H, W, 1) uint8 numpy array of binary images with values in {0, 1}
  image_shape: (H, W), height and width of the image
  dset_id: An identifying number of which dataset is given (1 or 2). Most likely
           used to set different hyperparameters for different datasets

  Returns
  - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
  - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
  - a numpy array of size (100, H, W, 1) of samples with values in {0, 1}
  """

  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
  print(f"Using device: {device}")

  # Convert to PyTorch tensors and normalize to [-1, 1]
  train_data = torch.tensor(train_data, dtype=torch.float32).permute(0, 3, 1, 2) * 2 - 1
  test_data = torch.tensor(test_data, dtype=torch.float32).permute(0, 3, 1, 2) * 2 - 1

  # Move data to GPU
  train_data = train_data.to(device)
  test_data = test_data.to(device)

  # Hyperparameters
  batch_size = 128
  learning_rate = 1e-3
  epochs = 10

  # Initialize model and optimizer
  model = PixelCNN(image_shape).to(device)
  optimizer = optim.Adam(model.parameters(), lr=learning_rate)

  # Tracking losses
  train_losses = []
  test_losses = []

  # Initial test loss
  def compute_loss(data):
      logits = model(data)
      return F.binary_cross_entropy_with_logits(logits, (data + 1) / 2)

  with torch.no_grad():
      test_loss = compute_loss(test_data)
      test_losses.append(test_loss.item())

  # Training loop
  for epoch in range(epochs):
      # Shuffle training data
      perm = torch.randperm(len(train_data))
      train_data_shuffled = train_data[perm]

      epoch_train_losses = []
      for i in range(0, len(train_data), batch_size):
          batch = train_data_shuffled[i:i+batch_size]

          # Zero gradients
          optimizer.zero_grad()

          # Compute loss
          logits = model(batch)
          loss = F.binary_cross_entropy_with_logits(logits, (batch + 1) / 2)
          loss.backward()

          # Update parameters
          optimizer.step()

          epoch_train_losses.append(loss.item())

      # Average train loss for the epoch
      train_losses.extend(epoch_train_losses)

      # Compute test loss after epoch
      with torch.no_grad():
          test_loss = compute_loss(test_data)
          test_losses.append(test_loss.item())

      # Print epoch progress
      print(f"Epoch {epoch+1}/{epochs}, Train Loss: {np.mean(epoch_train_losses):.4f}, Test Loss: {test_loss.item():.4f}")

  # Generate samples using ancestral sampling
  def generate_samples(num_samples):
      H, W = image_shape
      samples = torch.zeros(num_samples, 1, H, W, device=device)

      with torch.no_grad():
          # Generate pixels one by one
          for h in range(H):
              for w in range(W):
                  # Get logits for this pixel based on previous pixels
                  logits = model(samples)

                  # Extract logit for this specific pixel
                  pixel_logit = logits[:, :, h, w]

                  # Sample from Bernoulli distribution
                  pixel_probs = torch.sigmoid(pixel_logit)
                  samples[:, :, h, w] = (torch.rand_like(pixel_probs) < pixel_probs).float() * 2 - 1

      # Move samples back to CPU for numpy conversion
      return samples.cpu().numpy()

  # Generate 100 samples
  samples = generate_samples(100)

  # Reshape and convert samples back to [0, 1]
  samples = samples.reshape(100, image_shape[0], image_shape[1], 1)
  samples = (samples + 1) / 2

  # Convert to uint8
  samples = (samples * 255).astype(np.uint8)

  return np.array(train_losses), np.array(test_losses), samples
```

```python
class AutoregressiveMaskedConv2d(nn.Conv2d):
    """
    Class extending nn.Conv2d to use masks for groups.
    Found inspiration at: <a href="https://github.com/anordertoreclaim/PixelCNN/blob/master/pixelcnn/conv_layers.py"></a>
    """

    def __init__(self, mask_type, in_channels, out_channels, kernel_size, data_channels=3, padding=0):
        super(AutoregressiveMaskedConv2d, self).__init__(in_channels, out_channels, kernel_size, padding=padding)
        mask = np.zeros(self.weight.size(), dtype=np.float32)

        height, width = kernel_size, kernel_size

        mask[:, :, :height // 2, :] = 1
        mask[:, :, height // 2, :width // 2 + 1] = 1

        def color_mask(out_color, in_color):
            a = (np.arange(out_channels) % data_channels == out_color)[:, None]
            b = (np.arange(in_channels) % data_channels == in_color)[None, :]
            return a * b

        for out_channel in range(data_channels):
            for in_channel in range(out_channel + 1, data_channels):
                mask[color_mask(out_channel, in_channel), height // 2, width // 2] = 0

        if mask_type == 'A':  # Then set the center to be one
            for channel in range(data_channels):
                mask[color_mask(channel, channel), height // 2, width // 2] = 0

        self.register_buffer('mask', torch.from_numpy(mask))

    def forward(self, x):
        self.weight.data *= self.mask
        return super().forward(x)


class AutoregressiveResidualMaskedConv2d(nn.Module):
    """
    Residual Links between MaskedConv2d-layers
    As described in Figure 5 in "Pixel Recurrent Neural Networks" by Aaron van den Oord et. al.
    """

    def __init__(self, in_dim):
        super().__init__()
        self.net = nn.Sequential(
            AutoregressiveMaskedConv2d('B', in_dim, in_dim // 2, kernel_size=1, padding=0),
            nn.ReLU(),
            AutoregressiveMaskedConv2d('B', in_dim // 2, in_dim // 2, kernel_size=7, padding=3),
            nn.ReLU(),
            AutoregressiveMaskedConv2d('B', in_dim // 2, in_dim, kernel_size=1, padding=0),
            nn.ReLU())

    def forward(self, x):
        return self.net(x) + x


class AutoregressiveColorPixelRCNN(nn.Module):
    """
    Pixel R-CNN-class using residual blocks from "Pixel Recurrent Neural Networks" by Aaron van den Oord et. al.
    """

    def __init__(self, in_channels, out_channels, conv_filters):
        super().__init__()
        self.net = nn.Sequential(
            # A 7x7 A-type convolution with batch norm
            MaskedConv2d('A', in_channels, conv_filters, kernel_size=7, padding=3),
            nn.BatchNorm2d(conv_filters),
            nn.ReLU(),
            # 8 Residual B-type convolutons with batch norms
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveResidualMaskedConv2d(conv_filters), nn.BatchNorm2d(conv_filters), nn.ReLU(),
            AutoregressiveMaskedConv2d('B', conv_filters, out_channels, kernel_size=1, padding=0)).cuda()

    def forward(self, x):
        return self.net(x)
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=1024):
        super(PositionalEncoding, self).__init__()

        # Learned positional embeddings
        self.pos_embed = nn.Embedding(max_len, d_model)

    def forward(self, x):
        # Create position indices
        position = torch.arange(x.size(1), dtype=torch.long, device=x.device).unsqueeze(0)
        return self.pos_embed(position).expand_as(x)

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super(MultiHeadAttention, self).__init__()

        assert d_model % num_heads == 0, "d_model must be divisible by num_heads"

        self.d_model = d_model
        self.num_heads = num_heads
        self.head_dim = d_model // num_heads

        # Linear projections
        self.q_proj = nn.Linear(d_model, d_model)
        self.k_proj = nn.Linear(d_model, d_model)
        self.v_proj = nn.Linear(d_model, d_model)
        self.out_proj = nn.Linear(d_model, d_model)

    def forward(self, x, mask=None):
        batch_size, seq_len, _ = x.size()

        # Project and split into heads
        q = self.q_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        k = self.k_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        v = self.v_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)

        # Transpose for attention calculation
        q = q.transpose(1, 2)  # (batch, heads, seq_len, head_dim)
        k = k.transpose(1, 2)
        v = v.transpose(1, 2)

        # Compute attention scores
        scores = q @ k.transpose(-2, -1) / math.sqrt(self.head_dim)

        # Apply mask if provided
        if mask is not None:
            scores = scores.masked_fill(mask == 0, float('-inf'))

        # Softmax and attention
        attn = F.softmax(scores, dim=-1)
        out = attn @ v

        # Combine heads and project
        out = out.transpose(1, 2).contiguous().view(batch_size, seq_len, self.d_model)
        return self.out_proj(out)

class TransformerBlock(nn.Module):
    def __init__(self, d_model, num_heads):
        super(TransformerBlock, self).__init__()

        # Multi-head attention
        self.attn = MultiHeadAttention(d_model, num_heads)

        # Layer normalization
        self.ln1 = nn.LayerNorm(d_model)
        self.ln2 = nn.LayerNorm(d_model)

        # Feed-forward network
        self.ff = nn.Sequential(
            nn.Linear(d_model, 4 * d_model),
            nn.GELU(),
            nn.Linear(4 * d_model, d_model)
        )

    def forward(self, x, mask=None):
        # Self-attention with residual connection
        x = x + self.attn(self.ln1(x), mask)

        # Feed-forward with residual connection
        x = x + self.ff(self.ln2(x))

        return x

class AutoregressiveTransformer(nn.Module):
    def __init__(self, image_shape, d_model=128, num_heads=4, num_layers=2):
        super(AutoregressiveTransformer, self).__init__()

        H, W, C = image_shape
        self.image_shape = image_shape

        # Total number of pixels plus BOS token
        self.seq_len = H * W * C + 1

        # Embedding for pixels (including BOS token)
        self.token_embed = nn.Embedding(4, d_model)  # 2-bit images (0-3)

        # Positional encoding
        self.pos_encoding = PositionalEncoding(d_model)

        # Transformer blocks
        self.transformer_blocks = nn.ModuleList([
            TransformerBlock(d_model, num_heads) for _ in range(num_layers)
        ])

        # Final prediction layer
        self.predict = nn.Linear(d_model, 4)  # 4-way softmax

        # BOS token
        self.bos_token = nn.Parameter(torch.randn(1, d_model))

    def create_causal_mask(self, seq_len):
        # Create lower triangular mask to prevent attending to future tokens
        mask = torch.tril(torch.ones(seq_len, seq_len)).unsqueeze(0).unsqueeze(0)
        return mask.to(torch.bool)

    def forward(self, x):
        # Flatten image and add BOS token
        batch_size = x.size(0)

        # Prepare input sequence
        x_flat = x.view(batch_size, -1)

        # Prepend BOS token
        bos = torch.zeros(batch_size, 1, dtype=torch.long, device=x.device)
        x_seq = torch.cat([bos, x_flat], dim=1)

        # Embed tokens
        x_embed = self.token_embed(x_seq)

        # Add positional encoding
        x_embed += self.pos_encoding(x_embed)

        # Causal mask
        mask = self.create_causal_mask(x_seq.size(1))

        # Pass through transformer blocks
        for block in self.transformer_blocks:
            x_embed = block(x_embed, mask)

        # Predict next token probabilities
        logits = self.predict(x_embed[:, :-1, :])

        return logits

def train_transformer(train_data, test_data, image_shape, dset_id,
                      n_epochs=15, batch_size=32, lr=1e-3):
    # Convert data to torch tensors
    train_data = torch.from_numpy(train_data).long()
    test_data = torch.from_numpy(test_data).long()

    # Move to GPU if available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Create datasets and loaders
    train_dataset = torch.utils.data.TensorDataset(train_data)
    test_dataset = torch.utils.data.TensorDataset(test_data)

    train_loader = torch.utils.data.DataLoader(train_dataset,
                                               batch_size=batch_size,
                                               shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_dataset,
                                              batch_size=batch_size)

    # Initialize model
    model = AutoregressiveTransformer(image_shape).to(device)

    # Optimizer with warmup and cosine decay
    optimizer = optim.Adam(model.parameters(), lr=lr)
    total_steps = n_epochs * len(train_loader)
    warmup_steps = 1000

    def lr_lambda(current_step):
        # Warmup
        if current_step < warmup_steps:
            return current_step / warmup_steps

        # Cosine decay
        progress = (current_step - warmup_steps) / (total_steps - warmup_steps)
        return max(0.0, 0.5 * (1.0 + math.cos(math.pi * progress)))

    scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

    # Tracking losses
    train_losses = []
    test_losses = []

    # Initial test loss
    model.eval()
    test_loss = compute_test_loss(model, test_loader, device)
    test_losses.append(test_loss)

    # Training loop
    for epoch in range(n_epochs):
        model.train()
        epoch_train_losses = []

        for batch_idx, batch in enumerate(train_loader):
            x = batch[0].to(device)

            # Zero gradients
            optimizer.zero_grad()

            # Forward pass
            logits = model(x)

            # Compute loss (negative log-likelihood)
            loss = compute_loss(logits, x)

            # Backward pass and optimize
            loss.backward()
            optimizer.step()
            scheduler.step()

            # Record loss
            epoch_train_losses.append(loss.item())

            # Print progress
            if batch_idx % 10 == 0:
                print(f"Epoch {epoch+1}/{n_epochs}, Batch {batch_idx}/{len(train_loader)}, Loss: {loss.item():.4f}")

        # Average train loss for this epoch
        train_losses.extend(epoch_train_losses)

        # Compute test loss
        model.eval()
        test_loss = compute_test_loss(model, test_loader, device)
        test_losses.append(test_loss)

        print(f"Epoch {epoch+1} completed. Train Loss: {np.mean(epoch_train_losses):.4f}, Test Loss: {test_loss:.4f}")

    # Generate samples
    samples = generate_samples(model, image_shape, 100, device)

    return np.array(train_losses), np.array(test_losses), samples

def compute_loss(logits, x):
    # Flatten image
    B, _, = x.shape
    x_seq = x.view(B, -1)

    # Compute cross-entropy for each pixel
    loss = F.cross_entropy(logits.view(-1, 4), x_seq[:, 1:].reshape(-1), reduction='mean')
    return loss

def compute_test_loss(model, test_loader, device):
    total_loss = 0
    with torch.no_grad():
        for batch in test_loader:
            x = batch[0].to(device)
            logits = model(x)
            loss = compute_loss(logits, x)
            total_loss += loss.item()
    return total_loss / len(test_loader)

def generate_samples(model, image_shape, num_samples, device):
    H, W, C = image_shape
    model.eval()

    # Initialize samples with BOS token
    samples = torch.zeros(num_samples, H, W, C, dtype=torch.long).to(device)

    with torch.no_grad():
        for h in range(H):
            for w in range(W):
                for c in range(C):
                    # Generate current pixel
                    logits = model(samples)

                    # Probability distribution for current pixel
                    probs = F.softmax(logits[:, h*W*C + w*C + c], dim=1)

                    # Sample from distribution
                    samples[:, h, w, c] = torch.multinomial(probs, 1).squeeze()

    return samples.cpu().numpy()
```

## Part (c) K, V Caching for Improved Inference
You may have noticed that generation from the transformer is quite slow. Part of this is just due to the autoregressive nature. However, another part is due to some computational inefficiency. At each forward pass of the model, we are performing repeat computation of the past sequence. Specifically, we can cache the key and values at the multi attention layer to more quickly predict at each step.

In self-attention, a sequence is processed by generating three vectors for each element in the sequence: a Query (Q), a Key (K), and a Value (V). These vectors are then used to compute attention scores and subsequently the output of the attention layer.
Mathematically, this can be represented as:
 * For each index $i$, compute $Q_i$, $K_i$, $V_i$ for the current element
 * Retrieve $K_{<i}$ and $V_{<i}$ from the cache (where $<i$ denotes all indices before the current one)
 * Compute the attention output using $Q_i$, $[K_{<i}, K_i]$, $[V_{<i}, V_i]$
